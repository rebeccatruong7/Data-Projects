## Zillow Home Value Index (ZHVI)
Zillow Home Value Index (ZHVI): A measure of the typical home value and market changes across a given region and housing type. It reflects the typical value for homes in the 35th to 65th percentile range. Available as a smoothed, seasonally adjusted measure and as a raw measure.

Find the original table and more information about what ZHVI is and how it’s calculated on this [overview page](https://www.zillow.com/research/methodology-neural-zhvi-32128/).

Start dataset with ZHVI.

```python
# load zillow zip code dataset
zhvi = pd.read_csv(r"C:\[insert computer file path]Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv")
zhvi
```

```python
# City = "San Francisco"
sf_zhvi = zhvi[zhvi['City'] == 'San Francisco']
sf_zhvi
```

```python
# count number of zip codes 
sf_zhvi.shape[0]
```

```python
# confirm that everything is a zipcode
sum(sf_zhvi['RegionType']!='zip')
```

```python
# choose the most recent column 
sf_zhvi = sf_zhvi[['RegionName','2024-10-31']]
sf_zhvi
```

```python
# clean up dataset
sf_zhvi = sf_zhvi.rename(columns={"RegionName":"postcode",
                                                    "2024-10-31":"zhvi"}).\
                                    sort_values('zhvi', ascending=False).\
                                    reset_index(drop=True)

sf_zhvi
```

## American Community Survey: Median Household Income
Conducted by Census Bureau for American Community Survey, 5 year Estimates B19013, 2022.

Estimate Median household income in the past 12 months (in 2022 inflation-adjusted dollars).

Find the table [here](https://data.census.gov/table/ACSDT5Y2022.B19013?q=median%20household%20income&g=040XX00US06$8600000&d=ACS%205-Year%20Estimates%20Detailed%20Tables
). Click [here](https://www.census.gov/content/dam/Census/library/publications/2023/acs/acsbr-017.pdf
) for subject definitions. 

```python
# load median household income data from American Community Survey, 5 year estimate, 2022
med_household_income = pd.read_csv(R"C:\[insert computer file path]\ACSDT5Y2022.B19013_2024-11-14T194910\ACSDT5Y2022.B19013-Data.csv", 
                                   usecols=[1,2], skiprows=[1])
med_household_income
```

```python
# edit name to only include zipcodes
med_household_income['NAME'] = med_household_income['NAME'].str[-5:].astype(int)
med_household_income['NAME']
```

```python
med_household_income = med_household_income.rename(columns={'NAME':'postcode', 
                                                            'B19013_001E':'med_household_income'})
med_household_income 
```

```python
# only include SF zipcodes
    # match to ZHVI zipcodes
    
sf_postcode_wealth = pd.merge(med_household_income, sf_zhvi, on='postcode', how='inner')
sf_postcode_wealth 
# df_merged = pd.merge(df1, df2, on='key', how='inner')
```

## Zillow Observed Rent Index (ZORI)
Zillow Observed Rent Index (ZORI): A smoothed measure of the typical observed market rate rent across a given region. 

Find the table [here](https://www.zillow.com/research/data/). For more detailed information, you can refer to the [ZORI methodology](https://www.zillow.com/research/methodology-zori-repeat-rent-27092/).

```python
# load ZORI dataset
zori = pd.read_csv(r"C:\[insert computer file path]\Zip_zori_uc_sfrcondomfr_sm_month.csv")
zori
```

```python
# City = "San Francisco"
sf_zori = zori[zori['City'] == 'San Francisco']
sf_zori
```

```python
# count number of zip codes 
sf_zori.shape[0]
```

```python
# confirm that everything is a zipcode
sum(sf_zori['RegionType']!='zip')
```
