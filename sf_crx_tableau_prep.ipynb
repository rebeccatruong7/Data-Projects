## Add in geocode data to crx_places
Add two columns of latitude and longitude based on address data, using Google Cloud geocoding API.

```python
# Google Cloud API key
API_KEY = 'AIzaSyB0sY68dg0paVVmbghIND011rGxqOoEDWw'

# Function to clean addresses by removing suite/unit numbers
def clean_address(address):
    # Remove common suite/unit patterns (e.g., #15, Suite 100, Apt 20)
    cleaned_address = re.sub(r"(,?\s*(Suite|Apt|Unit|#)\s*\d+\b)", "", address, flags=re.IGNORECASE)
    return cleaned_address.strip()

# Function to get latitude and longitude from an address
def get_lat_lng(address):
    if not address:  # Check for empty or null addresses
        return None, None
    try:
        cleaned_address = clean_address(address)  # Clean the address
        url = f"https://maps.googleapis.com/maps/api/geocode/json?address={cleaned_address}&key={API_KEY}"
        response = requests.get(url, timeout=10)
        response_json = response.json()
        if response_json['status'] == 'OK':
            location = response_json['results'][0]['geometry']['location']
            return location['lat'], location['lng']
        else:
            print(f"API error for address '{cleaned_address}': {response_json['status']}")
            return None, None
    except requests.exceptions.RequestException as e:
        print(f"Request error for address '{address}': {e}")
        return None, None

# Apply the function to the DataFrame
crx_places[['latitude', 'longitude']] = crx_places['address'].apply(
    lambda x: pd.Series(get_lat_lng(x))
)

print(crx_places)
```
